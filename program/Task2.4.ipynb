任务 2.4 以产品购买结果为 x 轴、拜访客户的通话时长为 y 轴，绘制拜访
客户的通话时长箱线图，并在报告中对结果进行必要分析。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_style('whitegrid')
plt.rcParams['font.sans-serif'] = ['SimHei']
color = sns.color_palette()
```

```python
df = pd.read_excel('./result1_3.xlsx')
```

```python
df['duration']
```

```python
df['y']
```

```python
plt.figure(figsize=(10,5))#设置画布的尺寸
plt.title('拜访客户的通话时长箱线图',fontsize=20)#标题，并设定字号大小
labels = '0','1'
data = df.groupby('y')['duration'].unique()
plt.boxplot(data,labels=labels)
plt.xlabel('产品购买结果')
plt.ylabel('拜访客户的通话时长')
```
