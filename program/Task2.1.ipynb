```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import numpy as np
```

```python
data = pd.read_csv("./data/short-customer-data.csv")
```

```python
data
```

```python
for i in ['user_id','job','marital','education','default','housing','loan','contact','month','day_of_week','poutcome','y']:
    lb = LabelEncoder()
    data[i] = lb.fit_transform(data[i])
```

```python
sns.heatmap(data.corr())
```

```python
plt.figure(figsize=(400,300))
result = data.corr(method='pearson')
result.index = data.corr(method='pearson').index
mask = np.zeros_like(result, dtype=np.bool)   #å®šä¹‰ä¸€ä¸ªå¤§å°ä¸€è‡´å…¨ä¸ºé›¶çš„çŸ©é˜µ  ç”¨å¸ƒå°”ç±»åž‹è¦†ç›–åŽŸæ¥çš„ç±»åž‹
mask[np.triu_indices_from(result)]= True      #è¿”å›žçŸ©é˜µçš„ä¸Šä¸‰è§’ï¼Œå¹¶å°†å…¶è®¾ç½®ä¸ºtrue
plt.figure(figsize=(8,6))
# cmap = sns.choose_diverging_palette()
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(result,square=True, cmap=cmap,mask=mask,fmt='.2f',annot=True)
plt.savefig('E:/photo/ç¬¬ä¸‰é—®ç›¸å…³æ€§')
```
