# 删除异常值的列

```python
train['Age'] = train['Age'].replace('-',"Nan").replace('0','Nan').replace('-','Nan').replace('1','Nan')
train = train[train["Age"]!='Nan']
```

# 将数据进行替换

```python
train['Age'] = train['Age'].apply(lambda x:x.replace(' ',"").replace('岁','')).astype('int')
train
```
