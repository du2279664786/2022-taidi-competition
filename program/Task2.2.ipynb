```python
data = pd.read_csv("./data/short-customer-data.csv")
```

```python
data['flg'] = 0
data.head()
```

```python
def con_age(x):
    if (0<x)&(x<29):     # 青年
        return 0
    elif (29<=x)&(x<49):    # 中年
        return 1
    elif (49<x)&(x<60):    # 中老年
        return 2
    elif x>=60:            # 老年
        return 3
```

```python
no_data = data[data['y']=='no']
no_age = pd.DataFrame(columns=['年龄','no_占比'])
for i in no_data['age'].unique():
    zhanbi = round(no_data[no_data['age']==i].shape[0]/no_data.shape[0],4)
    no_age.loc[len(no_age.index)] = [i, zhanbi]
#     no_age.append(pd.DataFrame({'年龄':i,'占比':zhanbi}),ignore_index = True)
no_age
```

```python
yes_data = data[data['y']=='yes']
yes_age = pd.DataFrame(columns=['年龄','yes_占比'])
for i in yes_data['age'].unique():
    zhanbi = round(yes_data[yes_data['age']==i].shape[0]/yes_data.shape[0],4)
    yes_age.loc[len(yes_age.index)] = [i, zhanbi]
#     yes_age.append(pd.DataFrame({'年龄':i,'占比':zhanbi}),igyesre_index = True)
yes_age
```

```python
data_all = pd.merge(yes_age,no_age,on='年龄')
```
